```python
%matplotlib inline
import matplotlib.pyplot as plt
import re
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
from html.parser import HTMLParser

class MLStripper(HTMLParser):
    def __init__(self):
        self.reset()
        self.strict = False
        self.convert_charrefs= True
        self.fed = []
    def handle_data(self, d):
        self.fed.append(d)
    def get_data(self):
        return ''.join(self.fed)

def strip_tags(html):
    s = MLStripper()
    s.feed(html)
    return s.get_data()
```

```python
questions_quality["parsed_text"] = questions_quality["Questions"].map(lambda x: re.sub("[ ]+", " ", (re.sub("[^A-Za-z ]+", " ", re.sub("'", "", strip_tags(str(x).lower()).split("|")[0]))).strip()))
```

```python
questions_quality["word_count"] = questions_quality["parsed_text"].map(lambda x: len(x.split(" ")))
```

```python
questions_quality["character_count"] = questions_quality["parsed_text"].map(lambda x: len(x))
```

```python
dict_set = dictionary_words["word"].values
questions_quality["dict_score"] = questions_quality["parsed_text"].map(lambda x: (len(x.split(" ")) - len([word for word in x.split(" ") if word not in dict_set]))/float(len(x.split(" "))))
```

```python
word_min = questions_quality["word_count"].min()
word_max = questions_quality["word_count"].max()
questions_quality["word_normalized"] = questions_quality["word_count"].map(lambda x: float(x - word_min)/(word_max - word_min))
```

```python
char_min = questions_quality["character_count"].min()
char_max = questions_quality["character_count"].max()
questions_quality["char_normalized"] = questions_quality["character_count"].map(lambda x: float(x - char_min)/(char_max - char_min))
```
