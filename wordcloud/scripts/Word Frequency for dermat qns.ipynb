```python
import os
import re
from textblob import Word
from collections import Counter
import operator
```

```python
def readQuestionFile():
	f = open('twitter.txt', 'r')
	f1 = open('stop_words_prim.txt', 'r')
	fileVal = f.read()
	fileVal2 = f1.read()
	questions = re.sub("[^\w]", " ",  fileVal.lower()).split()
	stop_words = re.sub("[^\w]", " ",  fileVal2.lower()).split()
	words = []
	print len(questions)
	for i in range(0, len(questions)):
		flag = 1
		for j in range(0, len(stop_words)):
			if questions[i] == stop_words[j] or len(questions[i]) < 3:
# 			words.append(questions[i])
				flag = 0
				break
		if flag == 1:
			words.append(questions[i])
    
	counts = Counter(words)
	print len(words)
	print len(stop_words)
# 	print words
# 	print(counts)
# 	print "Question_length : ", len(words)
	return counts
```

```python
questions = readQuestionFile()
```
